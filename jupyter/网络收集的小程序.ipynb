```python
#获取笔趣阁小说的目录夜页面的地址
mulu_url='http://www.xbiquge.la/xiaoshuodaquan/'
#全部小声说汇总的页面地址
tmp=rq.get(mulu_url,headers=headers)
```

```python
tmp2=bp(tmp.text,'lxml')
```

```python
print(tmp2)
```

```python
tmp3=tmp2.find_all('ul')
```

```python
novel=[]
for i in tmp3[1:]:
    tmp4=i.find_all('a')
    for j in tmp4:
        nl={}
        nl['novel_name']=j.text
        nl['novel_addr']=j.attrs['href']
        novel.append(nl)
for i in novel:
    print(i)
```

```python
#上面已经获取了所有小说的目录地址
#下面根据任意目录地址来获得所有小说的章节地址
#自动生成小说的名称
tmp=rq.get('http://www.xbiquge.la/17/17008/',headers=headers).content
```

```python
novel_list=[]
for i in tmp3:
    k=re.findall('<dd><a href=(.*? )>.*?</dd>',i)
    if k==[]:
        pass
    else:
        novel_list.append('http://www.xbiquge.la'+k[0].replace("'",''))
for i in novel_list:
    print(i)
```
