```python
import pandas as pd
from mailmerge import MailMerge as mmg

df=pd.read_excel('d:/material/name.xlsx')
num=len(df['name'])

for i in range(num):
    name=df['name'][i]
    doc=mmg('d:/material/mod.docx')
    doc.merge(name=name)
    doc.write(f'd:/material/{name}.docx')
    
#上面的程序用来对模板的docx文件里面的占位符进行改写并生成新的文件
```

###2.xlwt主要是用来写excel文件

    wbk = xlwt.Workbook()

    sheet = wbk.add_sheet('sheet 1')

    sheet.write(0,1,'test text')#第0行第一列写入内容

###3.xlutils结合xlrd可以达到修改excel文件目的

    import xlrd

    from xlutils.copy import copy

    workbook = xlrd.open_workbook(u'有趣装逼每日数据及趋势.xls')

    workbooknew = copy(workbook)

    ws = workbooknew.get_sheet(0)

    ws.write(3, 0, 'changed!')

    workbooknew.save(u'有趣装逼每日数据及趋势copy.xls')
